```python
sp_stock_df = pd.read_csv("sp500_stock_data.csv")
sp_stock_df.head()
```

```python
sp_stock_df2 = sp_stock_df[["Date", "Adj Close"]]
sp_stock_df2
```

```python
sp_stock_df2.plot(x="Date", y = "Adj Close", rot = 60)
```

```python
## 21 is the month look back
## 252  is the year look back
## new column
sp_stock_df2["daily returns"] = sp_stock_df2["Adj Close"].pct_change(1)
sp_stock_df2.head()
```

```python
sp_stock_df2["daily returns"].plot()
```

```python
sp_stock_df2["cum returns"] = (1 + sp_stock_df2["daily returns"]).cumprod()
sp_stock_df2.head()
```

```python
sp_stock_df2.plot(x="Date",y = "cum returns")
```
